```python
import pandas as pd
import json
import os
from datasets import load_dataset
```

```python
# deepfashion-inshop

gt_folder = './data/DeepFashion-InShop/gt_query_doc/'
```

```python
token="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"
ds = load_dataset('Marqo/deepfashion-inshop_chinese', num_proc=os.cpu_count(), cache_dir='./cache/', token=token)['data']
ds_ori = load_dataset('Marqo/deepfashion-inshop', num_proc=os.cpu_count(), cache_dir='./cache/', token=token)['data']
```

```python
df = ds.to_pandas()
df = df.sort_values(by='ID', ascending=True)
df_ori = ds_ori.to_pandas()
del ds
del ds_ori
df.head()
```

```python
df = df.drop('image', axis=1)
df_ori = df_ori.drop('image', axis=1)
df_ori.head()
```

```python
gt_files = os.listdir(gt_folder)
gt_files
```

```python
with open(gt_folder + 'ground_truth_color-image+text.json', 'r') as f:
    
    gt_color_text = json.load(f)
gt_color_text.keys()
```

```python
gt_color_translated = dict()
translated_key_set = dict()
```

```python

for key, value in gt_color_text.items():
    if key not in translated_key_set:
        row_indices = df_ori.index[df_ori['color'] == key].tolist()[0]
        translated_key = df.loc[row_indices]['color']
        translated_key_set[key] = translated_key
    else:
        translated_key = translated_key_set[key]
    
    if translated_key in gt_color_translated:
        gt_color_translated[translated_key].update(value)
    else:
        gt_color_translated[translated_key] = value
    
    
```

```python
len(gt_color_text)
```

```python
len(gt_color_translated)
```

```python
with open(gt_folder + 'ground_truth_color-image+text_chinese.json', 'w', encoding='utf-8') as f:
    json.dump(gt_color_translated, f, ensure_ascii=False, indent=4)
```
